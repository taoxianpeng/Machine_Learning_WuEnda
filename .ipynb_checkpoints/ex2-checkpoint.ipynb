```python
def featureScaling(data:pd.DataFrame)->pd.DataFrame:
    data_f = data
    # ç¬¬0åˆ—åŒ…å«äº†1ï¼Œæ‰€ä»¥è¦ä»Žç¬¬ä¸€åˆ—å¼€å§‹
    for col in range(1,data.shape[1]):
        data_col = data_f.iloc[:,col]
        data_col = (data_col-data_col.min()) / (data_col.max() - data_col.min())
        data_f.iloc[:,col] = data_col
    return data_f
```

```python
def sigmoid(z):
    return 1/(1 + np.exp(-z))
```

```python
#é€»è¾‘å›žå½’ ä»£ä»·å‡½æ•°
def cost_function(theta,X,y) -> np.ndarray:
    first = (-y) * np.log(sigmoid(X @ theta))
    second = (1 - y)*np.log(1 - sigmoid(X @ theta))
    return np.mean(first - second)
```

```python
x = np.linspace(-10,10,20)
plt.plot(x,sigmoid(x), 'r.')
```

```python
# def get_theta(Xi,yi, a, iterator) -> np.ndarray:  # æ¢¯åº¦ä¸‹é™æ±‚theta
#     X = np.matrix(Xi)
#     y = np.matrix(yi).T
#     theta = np.matrix(np.zeros((X.shape[1],1)))  # åˆå§‹åŒ–ä¸€ä¸ª Î¸ ä¸´æ—¶çŸ©é˜µ(1,n)
#     cost = np.zeros(iterator)  # åˆå§‹åŒ–ä¸€ä¸ªndarrayï¼ŒåŒ…å«æ¯æ¬¡epochçš„cost
#     m = X.shape[0]  # æ ·æœ¬æ•°é‡m

#     for i in range(iterator):
#         # åˆ©ç”¨å‘é‡åŒ–ä¸€æ­¥æ±‚è§£
#         theta= theta - (a / m) * X.T * (sigmoid(theta, X) - y)

       
#         cost[i] = cost_funtion(theta,X,y)
#     return theta
```

```python
def gradient(theta, X, y):
    return (X.T @ (sigmoid(X @ theta) - y))/len(X)
```

```python
theta = np.zeros(X.shape[1])
```

```python
gradient(theta,X.values,y.values)
```

```python
# X_f = featureScaling(X)

Xi = X.values
yi = y.values
result = opt.fmin_tnc(func=cost_function, x0=theta, fprime=gradient, args=(Xi,yi))
```

```python
result
```

```python
#ä¸Žfmin_tncæ–¹æ³•ä¸€æ ·ï¼Œ minimizeå‡½æ•°å¯ä»¥ è®¾ç½®methodæ–¹æ³•è°ƒç”¨å…¶ä»–æ–¹æ³•ã€‚
res = opt.minimize(fun=cost_function, x0=theta, args=(Xi, yi), method='TNC', jac=gradient)
res
```

å¯¹æ±‚çš„thetaè¿›è¡Œæµ‹è¯•

```python
def test(x, theta):
    res = sigmoid(x @ theta)
    return [1 if n>= 0.5 else 0 for n in res]
```

```python
# x = np.array([1,30,40])
# theta_res = result

correct=[1 if a==b else 0 for (a,b) in zip(test(X.values,result[0]), y.values)]
accuracy = np.sum(correct) / len(correct)
accuracy
#å½“å‰æ¨¡åž‹çš„æ­£ç¡®çŽ‡
```

```python
def fearture_mapping(x1,x2,power):
    data = {"f{}{}".format(i-p,p) : np.power(x1,i-p)*np.power(x2,p) for i in np.arange(power+1) for p in np.arange(i+1)}
    return pd.DataFrame(data)
```

```python
data3 = fearture_mapping(df2['X0'].values,df2['X1'].values,6)
data3.head()
```

```python
data3.shape[1]
```

```python
def gradientReg(theta, X, y, a=1):
    #ä¸æƒ©ç½šç¬¬ä¸€ä¸ª
    m=X.shape[0]
    reg = (a/m) * theta
    reg[0] = 0
    return gradient(theta,X,y)+reg
```

```python
def costFunReg(theta,X,y,a=1):
    _theta = theta[1:] #ç¬¬ä¸€é¡¹ä¸æƒ©ç½š
    reg = (a/(2*len(X))) * (_theta @ _theta) #æ­£åˆ™å‚æ•°
    
    return cost_function(theta,X,y) + reg
```

```python
theta3 = np.zeros(data3.shape[1])
X3 = data3.values
y3 = df2['X2'].values
reslut2 = opt.fmin_tnc(func=costFunReg, x0=theta3,fprime=gradientReg,args=(X3, y3, 2))
#gradientReg(theta,X3,y3,2)
# theta3.shape,X3.shape, y3.shape
reslut2
```

```python
final_theta = reslut2[0]
predictions = test(X3,final_theta)
correct = [1 if a==b else 0 for (a, b) in zip(predictions, y3)]
accuracy = sum(correct) / len(correct)
accuracy
```
